```python
import torch
import numpy as np
import pandas as pd
```

```python
def create_windows(timeseries,lookback, horizon):
    X, y = [], []
    for i in range(len(timeseries) - lookback - horizon + 1):
        feature = timeseries[i:i+lookback]
        target = timeseries[i:i+lookback+horizon] # we want the target to contain the timesteps from the lookback AND the steps forward so that the lookback steps will be treated as features

        X.append(feature)
        y.append(target)

    X = torch.tensor(X)
    y = torch.tensor(y)

    return X, y

def ConstantMinMaxScaler(series,min_gold,max_gold):
    #print(type(series))
    #series = np.ndarray(series)
    dataset_scaled = (series - min_gold)/(max_gold - min_gold)
    return dataset_scaled

def ConstantUnScaler(series, min_gold, max_gold):
    #series = np.ndarray(series)
    dataset_unscaled = np.round((series * (max_gold - min_gold)) + min_gold)
    return dataset_unscaled


max_gold = np.float64(104209.0) #all_heroes_ts.max().max() - this comes from the df2 in
min_gold = 0
```
